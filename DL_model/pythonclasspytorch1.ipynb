================

Getting Started
---------------

Tensors
^^^^^^^

Tensors are similar to NumPy’s ndarrays, with the addition being that
Tensors can also be used on a GPU to accelerate computing.



```python
from __future__ import print_function
import torch
```

<div class="alert alert-info"><h4>Note</h4><p>An uninitialized matrix is declared,
    but does not contain definite known
    values before it is used. When an
    uninitialized matrix is created,
    whatever values were in the allocated
    memory at the time will appear as the initial values.</p></div>



Construct a 5x3 matrix, uninitialized:



```python
x = torch.empty(5, 3)
print(x)
```

Construct a randomly initialized matrix:



```python
x = torch.rand(5, 3)
print(x)
```

Construct a matrix filled zeros and of dtype long:



```python
x = torch.zeros(5, 3, dtype=torch.long)
print(x)
```

Construct a tensor directly from data:



```python
x = torch.tensor([5.5, 3])
print(x)
```

or create a tensor based on an existing tensor. These methods
will reuse properties of the input tensor, e.g. dtype, unless
new values are provided by user



```python
x = x.new_ones(5, 3, dtype=torch.double) 
print(x)   # 只继承原张量的设备和一些默认设置，但是大小和数据内容是你指定的
# 一般用来新建张量，只是懒得手动写 device、layout 这些信息

x = torch.randn_like(x, dtype=torch.float)    
print(x)   #自动继承原张量的大小、dtype、device。    

# new_* 方法（如 new_ones），你手动指定。	*_like 方法（如 randn_like）从原张量继承
```

Get its size:



```python
print(x.size())
```

<div class="alert alert-info"><h4>Note</h4><p>``torch.Size`` is in fact a tuple, so it supports all tuple operations.</p></div>

Operations
^^^^^^^^^^
There are multiple syntaxes for operations. In the following
example, we will take a look at the addition operation.

Addition: syntax 1



```python
y = torch.rand(5, 3)
print(x + y)
```

Addition: syntax 2



```python
print(torch.add(x, y))
```

Addition: providing an output tensor as argument



```python
result = torch.empty(5, 3)
torch.add(x, y, out=result)
print(result)
```

Addition: in-place



```python
# adds x to y   in-place操作
y.add_(x) 
print(y) #直接修改原始张量的值，不会返回新张量。
```

<div class="alert alert-info"><h4>Note</h4><p>Any operation that mutates a tensor in-place is post-fixed with an ``_``.
    For example: ``x.copy_(y)``, ``x.t_()``, will change ``x``.</p></div>

You can use standard NumPy-like indexing with all bells and whistles!



```python
print(x[:, 1])
```

Resizing: If you want to resize/reshape tensor, you can use ``torch.view``:



```python
x = torch.randn(4, 4)
y = x.view(16) #重塑（reshape） -1 表示“自动推断”。
z = x.view(-1, 8) 
print(x.size(), y.size(), z.size())
```

If you have a one element tensor, use ``.item()`` to get the value as a
Python number



```python
x = torch.randn(1)
print(x)
print(x.item()) #.item() 是 从单元素张量中提取出其纯 Python 数值
```

## NumPy 桥接

将一个 Torch 张量（Tensor）转换为 NumPy 数组，或者反过来转换，操作非常简单。

如果 Torch 张量在 **CPU** 上，Torch 张量和 NumPy 数组会**共享底层内存**，因此**修改其中一个，另一个也会跟着变化**。

### 将 Torch 张量转换为 NumPy 数组




```python
a = torch.ones(5)
print(a)
```

```python
b = a.numpy() #Tensor 在 CPU 上时，才能 .numpy()
print(b)
```

See how the numpy array changed in value.



```python
a.add_(1)
print(a)
print(b) #共享内存
```

Converting NumPy Array to Torch Tensor
^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^
See how changing the np array changed the Torch Tensor automatically



```python
import numpy as np
a = np.ones(5)
b = torch.from_numpy(a) #把 NumPy 数组 a 转成 PyTorch 张量 b。
np.add(a, 1, out=a) #这个张量 b 与 NumPy 数组 a 共享内存
print(a)
print(b)
```

All the Tensors on the CPU except a CharTensor support converting to
NumPy and back.

CUDA Tensors
------------

Tensors can be moved onto any device using the ``.to`` method.



```python
# 仅当 CUDA 可用时，运行这段代码
# 我们将使用 `torch.device` 对象，把张量在 CPU 和 GPU 之间移动

if torch.cuda.is_available():
    device = torch.device("cuda")          # 创建一个 CUDA 设备对象（GPU）
    y = torch.ones_like(x, device=device)  # 直接在 GPU 上创建一个和 x 同形状的全1张量
    x = x.to(device)                       # 把原本的 x 复制到 GPU 上
    z = x + y                              # 在 GPU 上进行加法运算
    print(z)
    print(z.to("cpu", torch.double))       # 把 z 移回 CPU，同时改成 double 类型
```
