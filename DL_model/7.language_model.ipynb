一个句子 $S = \{w_1, w_2, w_3, \dots, w_n\}$

这个句子在现实世界中发生的概率 $P(S)$链式法则：
$$P(S) = P(w_1) \cdot P(w_2|w_1) \cdot P(w_3|w_1, w_2) \dots P(w_n|w_1, \dots, w_{n-1})$$

评估指标：
1. 困惑度 (Perplexity, PPL)：衡量的是模型在预测下一个词时有多“迷茫”。
2. 交叉熵损失  优化目标通常是交叉熵损失 $PPL = e^{Loss}$
