```python
import torch.nn as nn

class SimpleNet(nn.Module):
    def __init__(self):
        super(SimpleNet, self).__init__()
        # 定义一层：输入 10 个特征，输出 5 个神经元
        self.fc1 = nn.Linear(10, 5)
        self.relu = nn.ReLU()
        # 输出层：5 进 1 出
        self.fc2 = nn.Linear(5, 1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
```
