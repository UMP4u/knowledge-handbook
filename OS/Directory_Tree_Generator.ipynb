# ==========================================
# tree-node-cli 工具使用速查手册
# ==========================================

# 1. 基本运行 (无需安装，直接使用 npx)
npx tree-node-cli 
(npx是包运行器)

# 2. 限制层级 (Level)
# 推荐知识库使用 -L 2 或 -L 3，避免输出过长
npx tree-node-cli -L 2

# 3. 忽略特定文件夹 (Ignore)
# 使用引号包裹，多个目录用 | 分隔
npx tree-node-cli -I ".git|node_modules|__pycache__|.ipynb_checkpoints"

# 4. 只显示目录 (Files Only)
# 如果只想看文件夹架构，不看具体文件，加上 -f
npx tree-node-cli -f

# 5. 常用组合命令 (最推荐用于 README)
# 解释：限制2层深度 + 排除杂质文件夹 + 包含所有文件
npx tree-node-cli -L 2 -I ".git|node_modules|.vscode" -a

# 6. 将结果保存到文件
npx tree-node-cli -L 3 -o project_structure.txt

# ==========================================
# 参数简表：
# -L : 层级深度
# -I : 忽略模式 (支持正则表达式)
# -a : 显示隐藏文件
# -f : 只展示目录
# -o : 输出到指定文件
# ==========================================
