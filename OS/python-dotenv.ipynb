# Python 项目环境管理笔记：python-dotenv

## 1. 核心作用
`.env` 文件（Environment Variables）的主要作用是 **“配置与代码分离”**：
- **安全性**：避免将敏感信息（如 Hugging Face Token、数据库密码）硬编码在代码中并上传到 GitHub。

---

## 2. 准备工作

### 安装依赖
```bash
pip install python-dotenv
```
### 加入.gitignore
```bash
.env
```

## 3.文件编写格式(根目录创建)
```bash
HF_TOKEN=hf_vPxx  # 使用大写字母和下划线;等号两边不要空格
        #简单字符串不需要引号
        #包含空格的值,必须加引号
```
## 4.Python 代码调用
```python
import os
from dotenv import load_dotenv

# 1. 加载 .env 文件中的变量到系统环境变量
load_dotenv()

# 2. 使用 os.getenv 读取变量
# 建议设置默认值，防止变量缺失导致程序崩溃
hf_token = os.getenv("HF_TOKEN")
model_path = os.getenv("MODEL_NAME", "distilbert-base-uncased")
num_labels = int(os.getenv("NUM_LABELS", 2)) # 注意：读取出的值全是字符串，需要转换类型

```
