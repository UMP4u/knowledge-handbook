```python
解压
tar -xzvf cifar-100-python.tar.gz


-x：Extract（解压/提取）
-z：通过 gzip 过滤（处理 .gz 后缀）
-v：Verbose（显示进度，看有哪些文件被解压出来了）
-f：File（指定目标文件名，这个参数必须放在最后，后面紧跟文件名）

压缩（打包）
tar -czvf backup.tar.gz /path/to/folder

-c：Create（创建新的压缩包）
```
