```python
.gitignore语法

1.忽略特定后缀：使用星号 *
*.pth：忽略所有 PyTorch 权重文件
*.log：忽略所有日志文件(子目录下也可以src/data/error.log)

2.忽略整个文件夹：文件夹名后加斜杠 /
data/：忽略根目录下 data 文件夹及其内部所有内容
**/temp/：忽略任何子目录下的 temp 文件夹
(**/logs/*.log)

3.例外规则：使用感叹号 !
*.jpg
!cat.jpg：忽略所有 jpg 图片，但除了 cat.jpg
```

```python
基础工作流

git status 查看当前工作区和暂存区的状态，分为3类：
    Changes to be committed (暂存区中的文件)：这些是已经通过 git add 标记为“即将提交”的文件，显示为绿色。
    Changes not staged for commit (工作区已修改但未暂存的文件)：这些是你在本地修改了，但还没执行 git add 的文件，显示为红色。
    Untracked files (未追踪文件)：这些是新创建的、Git 之前完全没管过的文件（比如你新加的文件），显示为红色。

git add <file>: 将文件添加到暂存区（Staging Area）
    git add file1.py file2.py README.md
    git add DL_model/results/   #整个文件夹
```

```python
分支管理 (Branching)

git branch: 查看所有本地分支。

git branch <name>: 创建新分支。

git checkout <name> 或 git switch <name>: 切换到指定分支。

git merge <name>: 将指定分支合并到当前分支。

git branch -d <name>: 删除指定分支。

git rebase 不要在公共分支上使用 Rebase;公共主分支（如 main/develop）：只用 merge

命名 develop / dev	开发分支
```

```python
版本回退与状态重置

git log: 查看提交历史记录(commit ID)
    git log --oneline 简洁模式，只显示一行

git reset --mixed <commit-id>: 移动 HEAD，同时清空暂存区，让文件回到“未暂存”状态.适合你想撤销提交，同时也撤销暂存区记录，但保留本地代码修改的情况。
git reset --hard <commit-id>: 强制将代码库回退到指定版本（之前的修改会丢失）。

git revert <commit-id>: 生成一个新的提交来撤销之前的某次提交（更安全，不删除历史）(团队协作)

git rm --cached <file>: 将文件从 Git 追踪中移除，放回暂存区，但保留本地磁盘文件。

git restore --staged file.txt   该文件会从已暂存状态退回到“未暂存（Changes not staged for commit）”状态
git restore --staged .  把暂存区的所有改动全部撤销
```
