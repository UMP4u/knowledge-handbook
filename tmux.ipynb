```python
================================================================================
TMUX 核心操作与层级指南 (Cheat Sheet)
================================================================================

1. 三大核心层级 (The Three Levels)
--------------------------------------------------------------------------------
[会话 Session] -> [窗口 Window] -> [面板 Pane]

- 会话 (Session): 最高层。包含多个窗口。断开连接后仍可在后台持久运行。
  (用途: 隔离不同项目，如“网站开发”和“服务器监控”)
  
- 窗口 (Window):  中层。类似浏览器的“标签页 (Tab)”，占据整个屏幕。
  (用途: 同一项目的不同任务，如 0号写代码, 1号查日志, 2号运行数据库)
  
- 面板 (Pane):    底层。将一个窗口切割成多个小格子。
  (用途: 实时对比，如左边修改配置，右边观察实时报错)

2. 核心快捷键 (默认前缀键: Ctrl + b)
--------------------------------------------------------------------------------
# 会话管理 (在 Shell 命令行执行)
$ tmux new -s <name>      # 创建并进入一个具名Session
$ tmux ls                 # 列出所有正在运行的会话
$ tmux a -t <name>        # 重新连接到指定会话
$ tmux kill-session -t <n># 彻底关闭某个会话

# 窗口控制 (在 Tmux 内部操作: Prefix + ...)
Prefix + c                # 创建新窗口 (Create)
Prefix + ,                # 重命名当前窗口 (Rename)
Prefix + [0-9]            # 按数字直接跳转窗口
Prefix + w                # 弹出窗口列表，交互式选择

# 导航与杂项
Prefix + [                # 进入“复制模式” (可翻页查看历史记录)
q (在复制模式下)           # 退出查看模式

3. 状态栏符号解析 (Status Bar Indicators)
--------------------------------------------------------------------------------
[1] 0:0.1case*  1:win2 2:bash-

[1]                      # 当前会话名称 (Session Name)
0:0.1case*               # 窗口索引 : 名字 (* 代表当前正在操作的窗口)
1:win2
2:bash-                  #  (- 代表上一个停留过的窗口)
================================================================================
```
