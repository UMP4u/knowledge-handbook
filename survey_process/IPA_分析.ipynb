```python
IPA(Importance-Performance Analysis,重要性-表现分析法)的
核心原理基于“满意度取决于预期的重要程度与实际表现的差距”这一认知心理学假设。

横轴 (Importance)   纵轴 (Performance)    形成四象限


高重要 + 高表现 ：企业必须维持现状的“核心优势”

高重要 + 低表现 ：这是“机会点”或“痛点”，必须立即改进

低重要 + 低表现 ：不紧急，属于“无关紧要的短板”，投入精力回报率低

低重要 + 高表现 ：属于“过度投入”，虽然做到了极致，但用户并不买账，可考虑削减成本投入到第二象限
```

```python
执行步骤：

第一步：数据采集（李克特量表）  
    针对每一个评估指标（如“环境”、“口感”），要求用户打两分：重要性评分 ($I_i$) 和表现评分 ($P_i$)

第二步：计算每个指标的均值  $\bar{I}_i$ 和 $\bar{P}_i$

第三步：确定切分线  将所有指标重要性&表现评分的均值作为切分线，划分四象限

第四步：绘图定位
```
