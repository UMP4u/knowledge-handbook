```r
# ==========================================================
# 蒙特卡罗法估算定积分
# 目标：计算 g(x) 在 [a, b] 上的积分(在矩形框里撒点的投点法)
# ==========================================================

# 1. 定义被积函数
g <- function(x) {
  return(x^2)  
}

# 2. 设定参数
a <- 0          # 区间左端点
b <- 2          # 区间右端点
M <- 4          # 函数在 [a, b] 上的最大值 (必须保证在该区间内所有 g(x) <= M)
n <- 1000000    # 模拟撒点的总次数 (n 越大越精确)

# 3. 产生随机点
x_rand <- runif(n, min = a, max = b)
y_rand <- runif(n, min = 0, max = M)

# 4. 统计“命中”次数 (即落在曲线下方的点)
k <- sum(y_rand <= g(x_rand))

# 5. 计算结果
total_area <- (b - a) * M
integral_hat <- total_area * (k / n)

# 6. 输出结果
cat("模拟次数:", n, "\n")
cat("积分估算值:", integral_hat, "\n")
cat("理论参考值:", 8/3, "\n")

# ==========================================================
# 💡 提示：
# 如果 g(x) 的最大值 M 不容易确定，可以先用 optimize() 函数寻找：
# M <- optimize(g, interval = c(a, b), maximum = TRUE)$objective
# ==========================================================
```

```r
# 法2：均值法

g <- function(x) x^2  # 示例函数
a <- 0
b <- 2
n <- 1000000

# 1. 生成 [0, 1] 的随机数 y 
y <- runif(n, 0, 1)

# 2. 映射回 [a, b] 区间的 x 
x_mapped <- a + (b - a) * y #y是归一化

# 3. 计算 h(y) 的均值
# 注意：公式里的 (b-a) 可以提到 sum 外面，即 (b-a) * mean(g(x))
I_hat <- (b - a) * mean(g(x_mapped))

print(I_hat)
```
