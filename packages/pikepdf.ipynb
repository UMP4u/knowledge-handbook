1. 批量处理权限与密码
2. 无损提取与替换图片



```python
# 批量解密pdf文件

from pathlib import Path
import pikepdf

# 1. 初始化路径对象
input_root = Path('./encrypted_pdfs')  # 源文件夹对象
output_root = Path('./decrypted_pdfs') # 输出文件夹对象
password = '你的统一密码'               # 密码

def batch_decrypt_pathlib():
    # 2. 递归遍历所有 PDF 文件
    # .rglob('*.pdf') 
    # 它会自动进入所有子文件夹寻找后缀为 .pdf 的文件（忽略大小写建议用 *.[pP][dD][fF]）
    for input_path in input_root.rglob('*.pdf'):
        try:
            # 3. 计算相对路径并构建输出路径
            # input_path.relative_to(input_root) 提取文件在源目录下的相对结构
            relative_path = input_path.relative_to(input_root)
            output_path = output_root / relative_path  # 使用 / 运算符直接拼接路径
            
            # 4. 创建必要的子文件夹
            output_path.parent.mkdir(parents=True, exist_ok=True)
            
            # 5. 执行解密并保存
            with pikepdf.open(input_path, password=password) as pdf:
                pdf.save(output_path)
                print(f"✅ 成功: {relative_path}")
                
        except pikepdf.PasswordError:
            print(f"❌ 密码错误: {input_path.name}")
        except Exception as e:
            print(f"⚠️ 无法处理 {input_path.name}: {e}")

if __name__ == "__main__":
    batch_decrypt_pathlib()
    print("\n--- 任务结束 ---")
```
