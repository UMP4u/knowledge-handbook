```python
import time
import random
from tqdm import tqdm

# 1. 模拟参数空间
mu1_space = [10, 20, 30]
mu2_space = [0.1, 1.0]
n_runs = 2
total_runs = len(mu1_space) * len(mu2_space) * n_runs

best_nmi = 0.0

# 2. 使用 with 语句初始化，total 告知总步数，以便计算百分比和剩余时间
with tqdm(total=total_runs, desc="Grid Search Progress", colour='green') as pbar:##给进度条实例起个名字叫 pbar，方便后面手动更新
    
    for m1 in mu1_space:
        for m2 in mu2_space:
            for run in range(n_runs):
                # 【动态更新左侧】: 显示当前测试的参数组合
                pbar.set_description(f"Config(μ1={m1}, μ2={m2})")
                
                time.sleep(0.2) 
                current_nmi = random.uniform(0.5, 0.9)
                
                # 【动态更新右侧】: 如果发现更高分数，更新后缀显示 (Postfix)
                if current_nmi > best_nmi:
                    best_nmi = current_nmi
                    pbar.set_postfix({"Best_NMI": f"{best_nmi:.4f}"})
                
                # 【手动步进】: 完成一次任务，进度条前进 1
                pbar.update(1)

print(f"\n搜索完成！最佳 NMI: {best_nmi:.4f}")        
```
